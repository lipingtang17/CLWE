```python
in_dict1_file = "/mnt/ssd-201-112-01/cpii.local/lptang/dictionaries/MUSE/en-et.txt"
in_dict2_file = "/mnt/ssd-201-112-01/cpii.local/lptang/dictionaries/MUSE/en-fi.txt"
out_dict1_file = "/mnt/ssd-201-112-01/cpii.local/lptang/dictionaries/MUSE/et-fi.txt"
out_dict2_file = "/mnt/ssd-201-112-01/cpii.local/lptang/dictionaries/MUSE/fi-et.txt"
```

```python
from collections import defaultdict
def generate_dict(in_dict_file):
    in_dict = defaultdict(list)
    with open(in_dict_file, "r", encoding="utf-8") as f:
        for line in f.readlines():
            # print(line)
            in_dict[line.split()[0]].append(line.strip().split()[1])
    return in_dict
```

```python
in_dict1 = generate_dict(in_dict1_file)
in_dict2 = generate_dict(in_dict2_file)
```

```python
def generate_file(in_dict1, in_dict2, out_dict1_file):
    out_dict1 = defaultdict(list)
    for key, values in in_dict1.items():
        if key in in_dict2:
            for value in values:
                # if value not in in_dict2[key] and value != key:
                out_dict1[value] = in_dict2[key]

    with open(out_dict1_file, "w", encoding="utf-8") as f:
        for key, values in out_dict1.items():
            for value in values:
                # if value != key:
                f.write(key + "\t" + value + "\n")
    return out_dict1
```

```python
out_dict1 = generate_file(in_dict1, in_dict2, out_dict1_file)
out_dict2 = generate_file(in_dict2, in_dict1, out_dict2_file)
```
